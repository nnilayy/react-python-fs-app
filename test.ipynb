```python
import json
import requests
import yaml

def fetch_openapi_json(url='http://localhost:8000/openapi.json'):
    response = requests.get(url)
    return response.json()

def convert_json_to_yaml(json_data):
    return yaml.dump(json_data, allow_unicode=True)

openapi_json = fetch_openapi_json()
openapi_yaml = convert_json_to_yaml(openapi_json)

# Print the YAML to console or write it to a file
print(openapi_yaml)
# or write to a file
with open('openapi.yaml', 'w') as file:
    file.write(openapi_yaml)
```
